```python
# Dependencies and Setup
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from ipyleaflet import Map, Heatmap
```

```python
#Read mortality rates csv
mort_rate = "IMR_mortality_rate_2019.csv"

mort_rates_df = pd.read_csv(mort_rate)

mort_rates_df.head(10)
```

```python
number_of_countries = len((mort_rates_df["Country Name"]).unique())
iso_code = len((mort_rates_df["ISO Code"]).unique())
#number_of_purchases = (purchase_data["Purchase ID"]).count()
#total_revenue = (purchase_data["Price"]).sum()

# Create data frame with obtained values
summary_dataframe = pd.DataFrame({"Number of Countries":[number_of_countries],
                           "ISO Codes":[iso_code]})

summary_dataframe.head()
```

```python
#Read education data csv
edu_data = "Education_Data.csv"

edu_data_df = pd.read_csv(edu_data)

edu_data_df.head(10)
```

```python
num_of_countries = len((edu_data_df["LOCATION"]).unique())
indicator = len((edu_data_df["INDICATOR"]).unique())
#number_of_purchases = (purchase_data["Purchase ID"]).count()
#total_revenue = (purchase_data["Price"]).sum()

# Create data frame with obtained values
summary_df = pd.DataFrame({"Number of Countries":[num_of_countries],
                           "Indicators":[indicator]})

summary_df.head()
```

```python
countries = edu_data_df["LOCATION"]
time = edu_data_df["TIME"]
value = edu_data_df["Value"]
final = pd.DataFrame({"ISO Code":countries,
                      "Time":time,
                       "Value":value})
final.head(20)
```

```python
final.shape[0]
```

```python
unique_time= final["Time"].unique()
unique_time
```

```python
edu_final = final.pivot(index="ISO Code", columns= "Time")["Value"]
edu_final.head(15)
```

```python
#number of countries
mort_rates_df['ISO Code'].count()
```

```python
#reading lat/lon csv
lat_lon="latlon.csv"

lat_lon_df=pd.read_csv(lat_lon, encoding="ISO-8859-1")

lat_lon_df.head()
```

```python
# Store latitude and longitude in locations
locations = lat_lon_df[["lat", "lon"]]
```

```python
lat_lon_df
```

```python
lat_lon_df['edu18'].mean(),lat_lon_df['edu18'].median()
```

```python
#mortality data heat map 2018

from ipyleaflet import Map, Heatmap

m = Map(center=(40, -100), zoom=3, dragging=True)

lats = []
lngs = []
rtgs = []

for index,row in lat_lon_df.iterrows():
    lat = row['lat']
    lng = row['lon']
    rtg = row['mort18']
    lats.append(lat)
    lngs.append(lng)
    rtgs.append(rtg)
    
    
marker_locs = []
i=0

for i in range(0,len(lats)):
    att_rtg = (rtgs[i])*250
    marker_loc =  lats[i],lngs[i],att_rtg
    marker_locs.append(marker_loc)
    i+=1

heatmap = Heatmap(
    locations=marker_locs,

    
)

# Change some attributes of the heatmap
heatmap.radius = 10
heatmap.min_opacity=.5
heatmap.blur = 0
heatmap.max = 1


heatmap.gradient = {0.5: 'green', 0.7:'yellow',  0.9: 'orange', 1: 'red'}
m.add_layer(heatmap);

m
```

```python
#education data heat map 2018

mp = Map(center=(40, -100), zoom=3, dragging=True)

lats = []
lngs = []
rtgs = []

for index,row in lat_lon_df.iterrows():
    lat = row['lat']
    lng = row['lon']
    rtg = row['edu18']
    lats.append(lat)
    lngs.append(lng)
    rtgs.append(rtg)
    
    
marker_locs = []
i=0

for i in range(0,len(lats)):
    att_rtg = (rtgs[i]-30)*250
    marker_loc =  lats[i],lngs[i],att_rtg
    marker_locs.append(marker_loc)
    i+=1

heatmap = Heatmap(
    locations=marker_locs,

    
)

# Change some attributes of the heatmap
heatmap.radius = 10
heatmap.min_opacity=.5
heatmap.blur = 0
heatmap.max = 1


heatmap.gradient = {0.5: 'red', 0.7:'orange',  0.9: 'yellow', 1: 'green'}

mp.add_layer(heatmap);

mp
```
